### 4. Exploratory Analysis

```python
data.head()
```

```python
df = data.groupby('Sub-Category').sum()
#lables = data['Sub-Category'].unique().sort()
df
```

```python
x = [sub for sub, data in data.groupby('Sub-Category')]
plt.bar(x, df['Quantity'])
plt.xticks(x, rotation=90)
plt.xlabel('Sub Category')
plt.ylabel('Number of items sold')
plt.show()
```

####  Binders were the most bought items among customers

------------------

```python
states = data.groupby(['State']).sum().sort_values('Sales', ascending = False)
states = states.head(10)
states
```

#### Highest number of sales come from California and lowest from Montana for Furniture

---------------------------------

```python
data['Calc_Profit'] = data['Profit']/data['Quantity']
data.head()
```

```python
profit = pd.DataFrame(data.groupby(['Category','Sub-Category']).sum()[['Calc_Profit']]).sort_values('Sub-Category')
profit.style.highlight_max(color = 'green', axis = 0)
```

```python
def neg(value):
    if type(value) != str and value < 0 :
        return 'color: red'
  
profit.style.applymap(neg)
```

```python
x = [sub for sub, data in data.groupby('Sub-Category')]
plt.bar(x, profit['Calc_Profit'])
plt.xticks(x, rotation=90)
plt.xlabel('Sub Category')
plt.ylabel('Profit per Item')
plt.show()
```

#### Highest Profits were made by copiers and the Items: Bookcases, Tables and Suplies made losses.

```python
data['Order Month'] = data['Order Date'].str.split('/').str[0]
data.head()
```

```python
df = data.groupby('Order Month').sum()
df
```

```python
x = [month for month, data in data.groupby('Order Month')]
plt.bar(x,df['Sales'])
plt.xlabel('Month')
plt.ylabel('Sales')
plt.show()
```

#### The most sales are in the month November and December, considering those are the festive months in USA.
#### More advertisements should be put up in the months of November, December, and September.
